```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(8)

diff = 0.3
time = 1000

data = []
for i in range(time):
    y_ = data[i-1] if data else 0
    data.append(y_ + np.random.uniform(-diff, diff))
    
x = np.arange(time).reshape(-1, 1)
y = np.array(data).reshape(-1, 1)
```

```python
plt.plot(x, data)
plt.show()
```

```python
train_size = 400
test_size = 100

x_tr, y_tr = x[-(test_size + train_size):-test_size], y[-(test_size + train_size):-test_size]
x_te, y_te = x[-test_size:], y[-test_size:]

len(y_tr), len(y_te)
```

```python
linear_reg = LinearRegression()
linear_reg.fit(x_tr, y_tr)
```

```python
pred_y = linear_reg.predict(x_te)
pred_y[:5]
```

```python
plt.plot(x_tr, y_tr)
plt.plot(x_te, y_te, alpha=0.3)
plt.plot(x_te, pred_y)
plt.show()
```
