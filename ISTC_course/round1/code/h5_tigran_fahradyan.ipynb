```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
data = []
class_count = 5
np.random.seed(123)
for i in range(class_count):
    data.append(np.random.multivariate_normal([np.random.uniform(0, 20), np.random.uniform(0, 20)],
                                              [[np.random.uniform(0, 5), np.random.uniform(-0.2, 0.5)],
                                               [np.random.uniform(-0.5, 0.8), np.random.uniform(0, 5)]],
                                              size=int(np.random.uniform(10, 500)))) 
```

```python
for index, datum in enumerate(data):
    plt.plot(datum[:, 0], datum[:, 1], "o", alpha=0.6, label=index)
plt.scatter(6, 0, c="cyan", marker="+", s=1000)
plt.legend();
```

```python
class NormalNativeBayes():
    
    def __init__(self):
        self.means = []
        self.cov = []
        self.p_c = []
    
    def train(self, train_data):
        num_points = sum([len(datum) for datum in data])
        for datum in train_data:
            self.means.append(np.mean(datum, axis=0))
            self.cov.append(np.cov(datum))
            self.p_c.append(datum.shape[0] / num_points)

    def predict(self, point):
        return [self.multivariate_normal.pdf(x, self.mean[i], self.cov[i])
                          for i in range(len(self.mean))]
```
