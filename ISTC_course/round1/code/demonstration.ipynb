## Polynomial Curve Fitting

```python
from numpy import array, sin, linspace, pi, random
```

```python
# data generation
X = linspace(0, 1, 15) * 2 * pi
data = sin(X) + random.normal(scale=0.15, size=len(X))
```

```python
plt.plot(X, sin(X), label="true function")
plt.plot(X, data, "bo", label="observations")
plt.legend();
```

```python
from numpy import poly1d, polyfit
```

```python
degree = 5  # change degree
y = poly1d(polyfit(X, data, degree))
plt.plot(X, y(X), label="M" + str(degree))
# plt.plot(X, sin(X))
plt.plot(X, data, "bo")
plt.legend()
plt.show()
print(y)
```

```python
plt.plot(range(1, max_m), train_error, "b", label="train")
plt.plot(range(1, max_m), test_error, "r", label="test")
plt.legend()
plt.show();
```
