```python
from scipy.stats import norm
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
```

**Normal dist**

```python
mean, var, skew, kurt = norm.stats(moments='mvsk')
```

```python
mean, var, skew, kurt
```

```python
x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
```

```python
plt.plot(x, norm.pdf(x))
```

```python
norm.pdf(x).max(), norm.pdf(x).min(), norm.pdf(x).sum()
```

**Guassian dist**

```python
N = 50
window = signal.gaussian(N, std=7)
plt.plot(np.linspace(-N/2, N/2, N), window)
```

**Multivariate**

```python
mean = [0, 0]
# cov = [[1, 0], [0, 100]]
cov = np.eye(2)

x, y = np.random.default_rng().multivariate_normal(mean, cov, 5000).T
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

Prob density function estimation with cov matrix.

```python
from scipy.stats import multivariate_normal

mean = [0, 0]
cov = np.eye(2)

x = np.random.normal(0, 10, (5, 2))
x = np.vstack((x, np.zeros((1, 2))))
y = multivariate_normal.pdf(x, mean=mean, cov=cov)
x, y
```
