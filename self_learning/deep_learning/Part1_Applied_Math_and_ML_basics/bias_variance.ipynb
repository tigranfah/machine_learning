## Bias and Variance.

**What is bias?**

Bias is the difference between the average prediction of our model and the correct value which we are trying to predict. Model with high bias pays very little attention to the training data and oversimplifies the model. It always leads to high error on training and test data.

**What is variance?**

Variance is the variability of model prediction for a given data point or a value which tells us spread of our data. Model with high variance pays a lot of attention to training data and does not generalize on the data which it hasn’t seen before. As a result, such models perform very well on training data but has high error rates on test data.

### Bias.

Recall: We will denote a point estimate of a parameter θ by θˆ. 

A point estimator or statistic is any function of the data:

$$
    θˆ_{m} = g(x^{(1)},...,x^{(m)}).
$$



The bias of an estimator is defined as:

$$
    bias(θˆ_{m}) = E(θˆ_{m}) − θ
$$

An estimator θˆ_m is said to be unbiased if bias(θˆ_m) = 0. Or asymptotically unbiased if:

$$
    \lim_{m→∞} bias(θˆ_{m}) = 0
$$

________________________________________________________________________________________

Consider the bernulii distribution.

Take sumples. with mean θ.
$$
    \{x^{(1)}, . . . , x^{(m)}\}
$$
$$
    P(x^{(i)}; θ) = θ^{x^{(i)}} (1 - θ)^{(1-x^{(i)})}
$$

A common estimator is the mean of all the samples. So,

$$
    θ_{m} = \frac{1}{m} \sum_{i=1}{x^{(i)}}
$$

To see if this estomator is biased we substitude this last equation into the bias equation.

### Variance and Standard Error

As we computed the bias using the expectation of the estimator, we can compute the variance.

$$
    Var(θ_{pred})
$$


Alternately, the square root of the variance is called the **standard error**, denoted SE(θˆ).

$$
    SE(u_{m}) = \sqrt{Var[\frac{1}{m} \sum_{i=1}{x^{(i)}}]} = \frac{σ^{2}}{\sqrt{m}}
$$

$$
    Var(x^{(i)}) = σ^2
$$

where σ^2 is the true variance of the samples x_i. Note that the standard error is often estimated by using an estimate of the actual σ.

*Example: Bernoulli Distribution.*

Consider samples generated from independently and identically from a Bernoulli distribution.
