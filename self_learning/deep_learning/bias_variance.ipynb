### Bias.

Recall: We will denote a point estimate of a parameter θ by θˆ. 

A point estimator or statistic is any function of the data:

$$
    θˆ_{m} = g(x^{(1)},...,x^{(m)}).
$$



The bias of an estimator is defined as:

$$
    bias(θˆ_{m}) = E(θˆ_{m}) − θ
$$

An estimator θˆ_m is said to be unbiased if bias(θˆ_m) = 0. Or asymptotically unbiased if:

$$
    \lim_{m→∞} bias(θˆ_{m}) = 0
$$

________________________________________________________________________________________

Consider the bernulii distribution.

Take sumples. with mean θ.
$$
    \{x^{(1)}, . . . , x^{(m)}\}
$$
$$
    P(x^{(i)}; θ) = θ^{x^{(i)}} (1 - θ)^{(1-x^{(i)})}
$$

A common estimator is the mean of all the samples. So,

$$
    θ_{m} = \frac{1}{m} \sum_{i=1}{x^{(i)}}
$$

To see if this estomator is biased we substitude this last equation into the bias equation.
